```python
import nibabel as nib
import skimage.io as io
import numpy as np
from skimage.transform import rotate
import matplotlib.pyplot as plt
img=nib.load('SMIR.Brain.XX.O.CT_4DPWI.346004.nii')
print('affine info printing:\n',img.affine)
header=img.header
print(header)
```

```python
img_arr=img.get_fdata()
img_arr=np.squeeze(img_arr)
print('img.arr',img_arr.shape)
fig, ax= plt.subplots(nrows=2,ncols=2)
plt.subplot(221)
io.imshow(rotate(img_arr[:,:,0,20],90))
plt.subplot(222)
io.imshow(rotate(img_arr[:,:,0,45],90))
plt.subplot(223)
io.imshow(rotate(img_arr[:,:,1,20],90))
plt.subplot(224)
io.imshow(rotate(img_arr[:,:,1,45],90))
plt.suptitle('raw data of 4DPWI') 
```
